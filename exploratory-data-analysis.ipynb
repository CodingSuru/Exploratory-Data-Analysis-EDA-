**Just counting the Experience Level according to the Experience**

```python
df.head(2)
```

```python
df['experience_level'].value_counts() #this code will show all the count different values of the experience_level columns.
```

**Data in Tabular form**

```python
df.groupby('experience_level')['Years_Experience']\
       .agg(['mean','count'])\
       .query('count >= 10')\
       .sort_values('mean')
```

**Data in graphical form**

```python
ax = df.groupby('experience_level')['Years_Experience']\
       .agg(['mean','count'])\
    .query('count >= 10')\
    .sort_values('mean')['mean']\
    .plot(kind='barh', figsize=(12, 5), title='Average Experience Level by Experience')
ax.set_xlabel('Experience Year')
ax.set_ylabel('Experience Level')
plt.show()
```
